```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars2.csv')
df.info()
```

```python
df.isna().mean()
```

```python
df.describe()
```

```python
brand = pd.DataFrame(df['Model'][i].split(" ")[0] for i in range(len(df['Model'])))
df.insert(0, 'Brand', brand)
df['Model'] = pd.DataFrame(' '.join(df['Model'][i].split(" ")[1:]) for i in range(len(df['Model'])))
df.head(20)
```

```python
df.columns = df.columns.str.lower()
```

```python
df.publishdate.unique()
```

```python
df['publishdate'].replace({'düýn': '15.06.2021','2 gün öň': '14.06.2021','3 gün öň': '13.06.2021' }, inplace=True)
df['publishdate'].replace(regex=['18 sag öň','5 sag öň','13 sag öň','11 sag öň','12 sag öň','16 sag öň','şu wagt','2 sag öň','3 sag öň','4 sag öň','6 sag öň','7 sag öň','8 sag öň','9 sag öň','10 sag öň','14 sag öň','15 sag öň','17 sag öň','19 sag öň','20 sag öň','21 sag öň','22 sag öň','23 sag öň'], value='16.06.2021', inplace=True)
df['publishdate'].replace(regex=['116.06.2021', '216.06.2021'], value='16.06.2021', inplace=True)
```

```python
df['publishdate'] = pd.to_datetime(df['publishdate'], dayfirst=True)
df.groupby('publishdate').agg('count').loc[:, 'year'].plot(kind='bar')
```

```python
df.engine = (df.engine) * 1000 #converting engine volume to cc
```

```python
# we wanna get how many days before every car published in the site
days =pd.DataFrame(str(pd.Timestamp(year=2021, month=6, day=17) - df.publishdate[i]).split(" ")[0] for i in range(len(df['publishdate'])))
df.insert(11, 'days', days)
df.days = df.days.astype(str).astype(int)
df
```

```python
# number of published car by year
df.year.value_counts().head()
```

```python
# price of cars by city
top_7 = df.groupby('city')['price'].sum().sort_values(ascending=False).head(7).reset_index()
others = df.groupby('city')['price'].sum().sort_values(ascending=False)[7:].reset_index().price.sum()

n_c = pd.DataFrame([['others', others]], columns=['city', 'price'])

top_7 = top_7.append(n_c, ignore_index=True)
x = top_7.city
y = top_7.price
# fig= plt.figure(figsize=(12,8))
fig, ax = plt.subplots(1,1,figsize=(8,8))
ax.pie(y, labels= x)
ax.axis('equal')
plt.show()
```

```python
#we have some mileage data as '-' and some those are so huge numbers those sounds unreal or written by mistake
#then we need to change the datatype from object to int and replace them to mean of dataframe 

#df.mileage =df.mileage.replace("-", np.nan)
#df.loc[df['mileage'] > 500000 , 'mileage'] = np.nan
df.mileage  =df[~df.mileage.isna()].mileage.astype(int)
df['mileage'] = df['mileage'].fillna(df.mileage.mean())
df.mileage.mean()
```

```python
cr_year = 2000
new_tb = df.query("(year >= @cr_year)&(model == 'Camry')& (engine != 0)")

new_tb = new_tb[new_tb['mileage'] < 50000]
sns.heatmap(new_tb.corr(), annot=True, fmt='.0%')
```

```python
cr_year = 2000
camry_cars = df.query("( mileage != '-') & (model == 'Camry') & (engine != 0)")
camry_cars = camry_cars.query("mileage < 50000")
from sklearn.preprocessing import LabelEncoder

for column in camry_cars.columns:
  if camry_cars[column].dtype == np.number:
    continue
  camry_cars[column] = LabelEncoder().fit_transform(camry_cars[column])

camry_cars = camry_cars.drop(['brand', 'model', 'publishdate', 'days'], axis=1)


plt.figure(figsize=(12,8))

sns.heatmap(camry_cars.corr(), annot=True, fmt='.0%')

```
